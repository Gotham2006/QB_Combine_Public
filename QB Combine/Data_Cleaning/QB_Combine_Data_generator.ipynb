```python
import pandas as pd
from pathlib import Path
final_df = []

path = Path("..") / "QB Combine" / "Data_Cleaning"
# Loop through years 2016 to 2021
for year in range(2016, 2022):
    
    filename = f"combine-results-{year}-QB.csv"
    df = pd.read_csv(filename)

    df = df[df[" PFF_for_first_4_years"].notna()] #Has a PFF grade
    final_df.append(df)

# Combine all dataframes
combined_df = pd.concat(final_df)
combined_df.to_csv("QB_Combine_Data.csv", index=False)
```
