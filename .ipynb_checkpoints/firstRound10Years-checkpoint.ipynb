```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

allDraftPicks = pd.read_csv("https://raw.githubusercontent.com/leesharpe/nfldata/master/data/draft_picks.csv")

draftPicks = allDraftPicks[(allDraftPicks["season"] >= 2014) & (allDraftPicks["round"] == 1)]

positionCounts = draftPicks.groupby(['season', 'position']).size().unstack(fill_value=0)

positionCounts.plot(kind='area', stacked=True, figsize=(10, 6), colormap='Spectral')

plt.title('First Round Picks by Position (2014 - 2024)', fontsize=12)
plt.ylabel('Number of Picks (Out of 32 Possible)', fontsize=12)
plt.xlabel('Year', fontsize=16)
plt.legend(title='Position', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
